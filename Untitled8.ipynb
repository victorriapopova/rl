```python
import os, random, time, json, warnings
import numpy as np
import matplotlib.pyplot as plt

import gymnasium as gym
from gymnasium.wrappers import RecordVideo
from stable_baselines3 import PPO, A2C
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy

warnings.filterwarnings("ignore")

SEED = 42
np.random.seed(SEED)
random.seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)

RESULTS_DIR = "results_track1"
VIDEO_DIR = os.path.join(RESULTS_DIR, "videos")
os.makedirs(VIDEO_DIR, exist_ok=True)

def make_env(seed=SEED):
    """Pendulum-v1 без Box2D, совместимо с SB3."""
    def _f():
        env = gym.make("Pendulum-v1")
        env.reset(seed=seed)
        env = Monitor(env)
        return env
    return _f

def moving_avg(x, k=50):
    if len(x) == 0:
        return x
    k = min(k, len(x))
    return np.convolve(x, np.ones(k)/k, mode="valid")

def load_monitor_rewards(monitor_file_path):
    rewards = []
    with open(monitor_file_path, "r") as f:
        for line in f:
            if line.startswith("{"):
                data = json.loads(line)
                if "r" in data:
                    rewards.append(data["r"])
    return rewards

print("Готово: Pendulum-v1, импорты и папки настроены.")
```

```python
# Векторный монитор и надёжные логи
import os, time, json, gc, csv
import numpy as np
import gymnasium as gym
from stable_baselines3 import PPO, A2C
from stable_baselines3.common.vec_env import DummyVecEnv, VecMonitor

RESULTS_DIR = "results_track1"
os.makedirs(RESULTS_DIR, exist_ok=True)

SEED = 42
TIMESTEPS = 150_000   # для быстрой проверки можно 75_000

configs = {
    "PPO_g095":  {"algo": "PPO", "gamma": 0.95},
    "PPO_g099":  {"algo": "PPO", "gamma": 0.99},
    "PPO_g0999": {"algo": "PPO", "gamma": 0.999},
    "A2C_g099":  {"algo": "A2C", "gamma": 0.99},
}

def make_env(seed=SEED):
    def _f():
        env = gym.make("Pendulum-v1")
        env.reset(seed=seed)
        return env
    return _f

def read_vecmonitor_rewards(monitor_csv_path):
    """Читает награды эпизодов из monitor.csv, который пишет VecMonitor."""
    rewards = []
    if not os.path.exists(monitor_csv_path):
        return rewards
    with open(monitor_csv_path, "r") as f:
        reader = csv.reader(f)
        for row in reader:
            # строки с метаданными начинаются с '#'
            if not row or row[0].startswith("#"):
                continue
            # формат: r,l,t, ... (reward, length, time)
            try:
                r = float(row[0])
                rewards.append(r)
            except Exception:
                pass
    return rewards

trained_paths = {}
histories = {}

for name, cfg in configs.items():
    print(f"\n=== Тренировка {name}: {cfg} ===")
    monitor_csv = os.path.join(RESULTS_DIR, f"{name}.monitor.csv")
    if os.path.exists(monitor_csv):
        os.remove(monitor_csv)

    # DummyVecEnv
    vec_env = DummyVecEnv([make_env(SEED)])
    # оборачиваем в VecMonitor и указываем filename => файл будет точно заполняться
    vec_env = VecMonitor(venv=vec_env, filename=monitor_csv)

    Model = PPO if cfg["algo"] == "PPO" else A2C
    model = Model("MlpPolicy", vec_env, seed=SEED, gamma=cfg["gamma"], verbose=0)

    t0 = time.time()
    model.learn(total_timesteps=TIMESTEPS)  # без progress_bar, чтобы не путать потоки
    print(f"Готово за {time.time()-t0:.1f} c")

    model_path = os.path.join(RESULTS_DIR, f"{name}.zip")
    model.save(model_path)
    trained_paths[name] = model_path


    vec_env.close()
    del model, vec_env
    gc.collect()

    # читаем готовый monitor.csv
    rewards = read_vecmonitor_rewards(monitor_csv)
    histories[name] = rewards
    print(f"Эпизодов в истории: {len(rewards)}  | лог: {monitor_csv}")

print("\nСохранённые модели:")
for k, v in trained_paths.items():
    print(" -", k, "→", v)

print("\nОтлично! Теперь запускай ШАГ 4 (графики/оценка/видео).")
```

```python
# ШАГ 4 — ГРАФИКИ, ОЦЕНКА И ВИДЕО
import os, csv, json, numpy as np, matplotlib.pyplot as plt, gymnasium as gym
from gymnasium.wrappers import RecordVideo
from stable_baselines3 import PPO, A2C
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy

RESULTS_DIR = "results_track1"
VIDEO_DIR = os.path.join(RESULTS_DIR, "videos")
os.makedirs(VIDEO_DIR, exist_ok=True)

names = ["PPO_g095","PPO_g099","PPO_g0999","A2C_g099"]
monitor_files = {n: os.path.join(RESULTS_DIR, f"{n}.monitor.csv") for n in names}
model_paths  = {n: os.path.join(RESULTS_DIR, f"{n}.zip")          for n in names}

def read_rewards(path):
    rews=[]
    with open(path,"r") as f:
        r=csv.reader(f)
        for row in r:
            if row and not row[0].startswith("#"):
                try: rews.append(float(row[0]))
                except: pass
    return rews

def moving_avg(x, k=50):
    if len(x)==0: return x
    k=min(k,len(x)); return np.convolve(x, np.ones(k)/k, mode="valid")

# 4.1 Кривые обучения
histories = {n: read_rewards(monitor_files[n]) for n in names}
plt.figure(figsize=(10,6))
for n in names:
    y = moving_avg(histories[n], k=50)
    x = np.arange(len(y))
    plt.plot(x, y, label=n)
plt.xlabel("Эпизод (скользящее окно=50)")
plt.ylabel("Суммарная награда за эпизод")
plt.title("Pendulum-v1 — кривая обучения")
plt.legend(); plt.grid(True); plt.show()

# 4.2 Количественная оценка (20 эпизодов)
eval_results = {}
for n in names:
    Model = PPO if n.startswith("PPO") else A2C
    def make_eval_env():
        e = gym.make("Pendulum-v1")
        return Monitor(e)
    vec = DummyVecEnv([make_eval_env])
    model = Model.load(model_paths[n], env=vec)
    m, s = evaluate_policy(model, vec, n_eval_episodes=20, render=False)
    eval_results[n] = (m, s)
    vec.close()

print("Итоговая оценка (20 эпизодов):")
for n,(m,s) in eval_results.items():
    print(f"{n}: mean={m:.2f} ± {s:.2f}")

# Барчарт
labels = list(eval_results.keys())
means  = [eval_results[k][0] for k in labels]
stds   = [eval_results[k][1] for k in labels]
plt.figure(figsize=(8,5))
plt.bar(range(len(labels)), means, yerr=stds)
plt.xticks(range(len(labels)), labels, rotation=15)
plt.ylabel("Средняя награда (20 эп.)")
plt.title("Pendulum-v1 — сравнение итоговых агентов")
plt.grid(True, axis="y"); plt.show()

# 4.3 Видео лучшего агента
best_name = max(eval_results, key=lambda k: eval_results[k][0])
print("Лучший агент:", best_name, "→", eval_results[best_name])
Model = PPO if best_name.startswith("PPO") else A2C
model = Model.load(model_paths[best_name])

env = gym.make("Pendulum-v1", render_mode="rgb_array")
env = RecordVideo(env, video_folder=VIDEO_DIR, episode_trigger=lambda e: True)
obs, info = env.reset(seed=42)
done = False; trunc = False
while not (done or trunc):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, trunc, info = env.step(action)
env.close()
print("Видео сохранено в:", VIDEO_DIR)
```

```python
# ПАТЧ ДЛЯ ВИДЕО: ручной рендер без RecordVideo и сброс last_u
import os, imageio, numpy as np, gymnasium as gym
from stable_baselines3 import PPO, A2C  # ✅ правильный импорт

RESULTS_DIR = "results_track1"
VIDEO_DIR = os.path.join(RESULTS_DIR, "videos")
os.makedirs(VIDEO_DIR, exist_ok=True)

# если best_name и model_paths не остались в памяти, пропиши вручную:
best_name = "PPO_g095"
model_paths = {best_name: os.path.join(RESULTS_DIR, f"{best_name}.zip")}

Model = PPO if best_name.startswith("PPO") else A2C
model = Model.load(model_paths[best_name])

env = gym.make("Pendulum-v1", render_mode="rgb_array")
obs, info = env.reset(seed=42)

frames = []
max_steps = 400  # достаточно на один эпизод
for _ in range(max_steps):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, trunc, info = env.step(action)
    # ключевой фикс: отключаем last_u, чтобы не было ошибки масштабирования
    try:
        env.unwrapped.last_u = None
    except Exception:
        pass
    frame = env.render()
    if frame is not None:
        frames.append(np.asarray(frame, dtype=np.uint8))
    if done or trunc:
        break

env.close()

# Сохранение MP4
mp4_path = os.path.join(VIDEO_DIR, f"{best_name}_episode.mp4")
with imageio.get_writer(mp4_path, fps=30, codec="libx264") as writer:
    for f in frames:
        writer.append_data(f)

print("Видео сохранено в:", mp4_path)
```

# Track 1 — Обучение агента в классической среде (Pendulum-v1)

## Цель

Продемонстрировать применение алгоритмов обучения с подкреплением для решения стандартной задачи Pendulum-v1 с использованием библиотеки stable-baselines3. Провести два контролируемых эксперимента и проанализировать их влияние на обучение.

## Среда

Среда Pendulum-v1 из Gymnasium. Задача агента — удерживать маятник в вертикальном положении, минимизируя отклонение и скорость вращения при помощи непрерывного управления моментом силы.

## Эксперименты

1. Влияние коэффициента дисконтирования γ на обучение алгоритма PPO (γ = 0.95, 0.99, 0.999).
   Гипотеза: при увеличении γ агент будет планировать на более длинный горизонт, но обучение станет менее стабильным.

2. Сравнение алгоритмов PPO и A2C при одинаковом γ=0.99.
   Гипотеза: PPO обеспечит более стабильную динамику и лучшую итоговую награду.

## Параметры обучения

* Алгоритмы: PPO, A2C
* Политика: MlpPolicy (2 слоя × 64 нейрона)
* Количество шагов обучения: 150 000
* Gamma: 0.95, 0.99, 0.999
* Seed: 42
* Среда: Pendulum-v1

## Результаты

Лучшие результаты показал агент PPO с γ=0.95.

Меньшее значение γ (0.95) дало наилучший результат: агент быстрее стабилизировался и достиг более высокой награды.
